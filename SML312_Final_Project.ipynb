```python
import geopandas as gpd

# Load community district boundaries shapefile -->this will be combined with coordinates to determine the district where each trip originated
cd_shapefile = '/Users/brookemccarthy/Desktop/SML312_2/Final Project /Community Districts/geo_export_4f49eb4f-b5e1-4edd-ac96-69e1c6f448f7.shp' #https://automating-gis-processes.github.io/CSC/notebooks/L2/geopandas-basics.html

# make into a geodataframe 
cd_gdf = gpd.read_file(cd_shapefile)

# set coordinate reference system 
cd_gdf.set_crs('EPSG:4326', allow_override=True, inplace=True) #https://geopandas.org/en/stable/docs/reference/api/geopandas.GeoDataFrame.set_crs.html

# inspect shapefile 
print(cd_gdf.head())
```

```python
from shapely.geometry import Point

# create geometry columns for starting and ending coordinates for january --> testing workflow for combined dataframe 
citi_bike_jan_18['start_geometry'] = citi_bike_jan_18.apply(lambda row: Point(row['start station longitude'], row['start station latitude']), axis=1) #https://shapely.readthedocs.io/en/2.0.6/reference/shapely.Point.html
citi_bike_jan_18['end_geometry'] = citi_bike_jan_18.apply(lambda row: Point(row['end station longitude'], row['end station latitude']), axis=1)

# convert to GeoDataFrame
gdf_jan_18 = gpd.GeoDataFrame(citi_bike_jan_18, geometry='start_geometry') #https://geopandas.org/en/v0.4.0/gallery/create_geopandas_from_pandas.html

# set coordinate reference system 
gdf_jan_18.set_crs('EPSG:4326', allow_override=True, inplace=True) #https://geopandas.org/en/stable/docs/reference/api/geopandas.GeoDataFrame.set_crs.html

print(gdf_jan_18.head())
```

```python
# combine the shapefile with community districts with the gdf for the citibike data 
# how=left means the join will keep all rows from citibike 
# predicate = within means gdf will be joined to a community district if it is within the district 
gdf_jan_18 = gpd.sjoin(gdf_jan_18, cd_gdf, how="left", predicate='within') #https://geopandas.org/en/stable/docs/user_guide/mergingdata.html

# create df for end coordinates and join
gdf_end_jan_18 = gpd.GeoDataFrame(citi_bike_jan_18, geometry='end_geometry')
gdf_end_jan_18.set_crs('EPSG:4326', allow_override=True, inplace=True)
gdf_end_jan_18 = gpd.sjoin(gdf_end_jan_18, cd_gdf, how="left", predicate='within')

# add end df to first one 
# boro_cd represents the community district 
gdf_jan_18['end_cd'] = gdf_end_jan_18['boro_cd']
```

```python
# inspect combined geodataframe 
print(gdf_jan_18[['start station latitude', 'start station longitude', 'end station latitude', 'end station longitude', 'boro_cd', 'end_cd']])
```

```python
# count trips starting in each CD
start_counts = gdf_jan_18['boro_cd'].value_counts().reset_index()
start_counts.columns = ['boro_cd', 'start_count']

# count trips ending in each CD 
end_counts = gdf_jan_18['end_cd'].value_counts().reset_index()
end_counts.columns = ['boro_cd', 'end_count']

# combine counts into one df
usage_counts = pd.merge(start_counts, end_counts, on='boro_cd', how='outer') #https://pandas.pydata.org/docs/reference/api/pandas.merge.html#pandas.merge

# find total bike usage in each CD
usage_counts['total_bike_usage'] = usage_counts['start_count'] + usage_counts['end_count']

print(usage_counts)
```

```python
# merge the usage counts back with the community district shapefile to make a heatmap 
cd_gdf_jan_18 = gpd.read_file('/Users/brookemccarthy/Desktop/SML312_2/Final Project /Community Districts/geo_export_4f49eb4f-b5e1-4edd-ac96-69e1c6f448f7.shp')
cd_gdf_jan_18 = cd_gdf.merge(usage_counts, how='left', on='boro_cd')
# inspect merged file 
print(cd_gdf_jan_18.head())
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm

# set axes and turn off grid 
fig, ax = plt.subplots(figsize=(10, 10))
ax.set_axis_off()

# fix color scale for easier comparisons across months --> values were found by plotting several months and finding the max and min across them
vmin_value = 6.0  
vmax_value = 558599.0 

# Plot heatmap of bike usage in CDs
cd_gdf_jan_18.plot(column="total_bike_usage", ax=ax, legend=True,
            cmap="OrRd", vmin=vmin_value, vmax=vmax_value) # https://geopandas.org/en/stable/docs/user_guide/mapping.html


ax.set_title("CitiBike Usage by Community District (January 2018)", fontsize=15)
plt.show()
```
